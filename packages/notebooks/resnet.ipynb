```python
import torchvision
```

```python
from torch import nn
from torchinfo import summary

weights_resnet18 = torchvision.models.ResNet18_Weights.DEFAULT
ResNet18 = torchvision.models.resnet18(weights=weights_resnet18)
for param in ResNet18.parameters():
    param.requires_grad = False


ResNet18.fc = nn.Linear(512, 2)
summary(ResNet18, input_size=(1, 3, 224, 224))
```

```python
from torch import nn
from torchinfo import summary

weights_resnet34 = torchvision.models.ResNet34_Weights.DEFAULT
ResNet34 = torchvision.models.resnet34(weights=weights_resnet34)
for param in ResNet34.parameters():
    param.requires_grad = False

ResNet34.fc = nn.Linear(ResNet34.fc.in_features, 2)
summary(ResNet34, input_size=(1, 3, 224, 224))
```
